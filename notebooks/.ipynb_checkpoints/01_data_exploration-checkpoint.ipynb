### 5. The meaning of each column

- Attrition_Flag: Whether the customer churned (Target variable)
- Customer_Age: Age of the customer (in years)
- Gender: Gender (M = Male, F = Female)
- Dependent_count: Number of dependents
- Education_Level: Educational Qualification of the account holder (example: high school, college graduate, etc.)
- Marital_Status: Married, Single, Divorced, Unknown
- Income_Category: Annual Income Category of the account holder (< $40K, $40K - 60K, $60K - $80K, $80K-$120K, > $120K, Unknown)
- Card_Category: Type of Card (Blue, Silver, Gold, Platinum)
- Total_Relationship_Count: Total number of bank products the customer uses
- Months_Inactive_12_mon: Number of inactive months in the last 12 months
- Contacts_Count_12_mon: Number of contacts with the bank in the last 12 months
- Credit_Limit: Credit Limit on the Credit Card
- Total_Revolving_Bal: Total Revolving Balance on the Credit Card
- Total_Trans_Amt: Total Transaction Amount (Last 12 months)
- Total_Trans_Ct: Total Transaction Count (Last 12 months)
- Total_Ct_Chng_Q4_Q1: Change in Transaction Count (Q4 over Q1)
- Avg_Utilization_Ratio: Average Card Utilization Ratio

### 7. The percentage of missing values

```python
unknown_mask = (X_cat == "Unknown") | (X_cat == "")
missing_percent = np.mean(unknown_mask, axis=0) * 100
for i, name in enumerate(cat_names):
    print(f"{name:30} Missing (Unknown): {missing_percent[i]:.2f}%")

plot_bar(cat_names, missing_percent, title='Missing Values Percentage by Categorical Feature', xlabel='Feature', ylabel='% Missing')
plt.show()
```

There are 3 columns having missing values

### 8. Numerical column: values distributed, min/max

```python
mins = X_num.min(axis=0)
maxs = X_num.max(axis=0)
means = X_num.mean(axis=0)
stds = X_num.std(axis=0)
for i, name in enumerate(num_names):
    print(f"{name:30} Min: {mins[i]:.2f}, Max: {maxs[i]:.2f}, Mean: {means[i]:.2f}, Std: {stds[i]:.2f}")
```

### 9. Categorical columns: values distributed, number of different values 

```python
for i in range(X_cat.shape[1]):
    unique, counts = np.unique(X_cat[:, i], return_counts=True)
    sorted_idx = np.argsort(-counts)
    print(f"\n* {cat_names[i]}")
    for val, cnt in zip(unique[sorted_idx], counts[sorted_idx]):
        print(f"  {val:25} {cnt:5} ({cnt/n_rows*100:.2f}%)")
```

### 10. Correlation between numeric columns

```python
corr = np.corrcoef(X_num.T)
plot_heatmap(corr, num_names)
plt.show()
```

### 11. Compare distributions by Attrition Flag

```python
churn_mask = y == 1
mean_attrited = X_num[churn_mask].mean(axis=0)
mean_existing = X_num[~churn_mask].mean(axis=0)
diff = mean_attrited - mean_existing
for i, name in enumerate(num_names):
    print(f"{name:30} Attrited Mean: {mean_attrited[i]:.2f}, Existing Mean: {mean_existing[i]:.2f}, Diff: {diff[i]:.2f}")
```

### 12. Outlier detection (IQR)

```python
q1 = np.percentile(X_num, 25, axis=0)
q3 = np.percentile(X_num, 75, axis=0)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers = np.mean((X_num < lower) | (X_num > upper), axis=0) * 100
for i, name in enumerate(num_names):
    print(f"{name:30} Outliers: {outliers[i]:.2f}%")
```
